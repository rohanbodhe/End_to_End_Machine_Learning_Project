## Cleaning Ticket Sales Data

```python
tickets.shape
```

```python
tickets.columns
```

```python
tickets.info()
```

```python
# converting  date columns to datetime format
tickets['event_date'] = pd.to_datetime(tickets['event_date'])
tickets['sale_date'] = pd.to_datetime(tickets['sale_date'])
```

```python
# Removing Duplicates
initial_count = len(tickets)
tickets = tickets.drop_duplicates()
final_count = len(tickets)
print(f"Removed {initial_count - final_count} duplicate records.")
```

```python
# Handling Missing Values
tickets.isnull().sum()
```

```python
# Price column 
tickets['ticket_price'].describe()
```

```python
# standardizing text columns 
tickets['event_type'] = tickets['event_type'].str.lower().str.strip()
tickets['city'] = tickets['city'].str.lower().str.strip()
tickets['seat_section'] = tickets['seat_section'].str.lower().str.strip()
```

```python
# days until event
tickets['days_until_event'] = (tickets['event_date'] - tickets['sale_date']).dt.days
tickets['days_until_event'].describe()
```

```python
# Adding derived columns 
tickets['revenue'] = tickets['ticket_price'] * tickets['quantity_sold']
tickets['event_month'] = tickets['event_date'].dt.month
tickets['event_dayofweek'] = tickets['event_date'].dt.dayofweek
tickets['is_weekend'] = tickets['event_dayofweek'].isin([5,6]).astype(int)
```
